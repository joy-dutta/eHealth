# Normal Data


## Creating Normal Data based on Domain Knowledge (Taking Rest) -TYPE 1

```python
#[          ecg,    bioimpedence,speed, distance, spo2, heart_attack]

min_max = ([60, 100], [12, 18], [0, 3.0], [0, 2.0], [95, 98.0],[0,1])


def make_series(low, high, name):
    if any(isinstance(_, float) for _ in (low, high)):
        func = np.random.uniform
    else:
        func = np.random.randint
    return pd.Series(func(low, high, size=(20000,)), name=name)

column_names = ("ecg", "bioimpedence", "speed", "distance", "spo2", "heart_attack")

df0_1 = pd.concat([make_series(lo, hi, name) for (lo, hi), name in zip(min_max, column_names)],
          axis=1)
```

```python
df0_1
```
